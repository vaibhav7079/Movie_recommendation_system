## Data Preparation

```python
def remove_space(entry):
    return str(entry).replace(" ","")
```

```python
df1['genres'] = df1['genres'].apply(remove_space)
df1['production_companies'] = df1['production_companies'].apply(remove_space)
```

```python
df1.head(3)
```

```python
df1.dropna(inplace = True)
```

```python
df1['tag']=df1['genres']  + df1['overview'] + df1['production_companies']
```

```python
df2=df1[['original_title','tag']]
df2.head(3)
```

```python
def clean_text(text):
    import spacy
    nlp=spacy.load("en_core_web_sm")
    doc=nlp(text)
    clean_text=[]
    for token in doc:
        if not token.is_stop and not token.is_left_punct and not token.is_space:
            clean_text.append(token.lemma_.lower())
    return " ".join(clean_text)
```

```python
df2['clean_text'] =df2['tag'].parallel_apply(clean_text)
```

```python
df2
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
vactor_array = CountVectorizer(max_features=2000).fit_transform(df2['clean_text']).toarray()
```

#### Similarity matrix

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vactor_array)
similarity
```

```python
similarity.shape
```

```python
similarity_df=pd.DataFrame(similarity,index = df2.original_title,columns=df2.original_title)
```

```python
similarity_df.head()
```

```python
similarity_df['Inception'].sort_values(ascending=False)[1:5]
```

```python
def recommend(movie_name,count):
    
    return list(similarity_df[movie_name].sort_values(ascending=False)[1:count+1].index.values)
```

```python
recommend("The Dark Knight",10)
```
